```python
arq = os.listdir('C:\\Users\\Antonio\\Desktop\\xpe\\Arquiteto de big-data\\M2\\Pratico\\BASE_TP\\BASE_TP')
dados = 'C:\\Users\\Antonio\\Desktop\\xpe\\Arquiteto de big-data\\M2\\Pratico\\BASE_TP\\BASE_TP\\'+arq.pop()
result = pd.read_csv(dados,sep='|',encoding='latin-1')

for x in arq:
    dados = 'C:\\Users\\Antonio\\Desktop\\xpe\\Arquiteto de big-data\\M2\\Pratico\\BASE_TP\\BASE_TP\\'+x
    df = pd.read_csv(dados,sep='|',encoding='latin-1')
    result = pd.concat([result, df], ignore_index=True)
df = result
df['data_coleta'] = pd.to_datetime(df['data_coleta'])
df['data_nascimento'] = pd.to_datetime(df['data_nascimento'])
```
